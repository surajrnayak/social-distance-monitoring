```python
# -*- coding: utf-8 -*-

# Form implementation generated from reading ui file 'C:\Users\SURAJ\Desktop\sdmm.ui'
#
# Created by: PyQt5 UI code generator 5.15.2
#
# WARNING: Any manual changes made to this file will be lost when pyuic5 is
# run again.  Do not edit this file unless you know what you are doing.


import sys
import cv2
import datetime
from PyQt5.QtGui import *
from PyQt5.QtWidgets import *
from PyQt5.QtCore import *
from PyQt5.uic import *
from PyQt5 import QtCore, QtGui, QtWidgets
import os
import os.path

from datetime import datetime  
from datetime import timedelta 

n=1
together=0
start2=datetime.now()
class Ui_Form(object):
    def setupUi(self, Form):
        Form.setObjectName("Form")
        Form.resize(1403, 893)
        Form.setStyleSheet("background-color: rgb(104, 104, 104);")
        self.label = QtWidgets.QLabel(Form)
        self.label.setGeometry(QtCore.QRect(90, -10, 1401, 131))
        self.label.setStyleSheet("color:rgb(255, 255, 255);\n"
"font: 75 italic 46pt \"Cambria\";\n"
"background-color: rgb(0, 0, 0);")
        self.label.setAlignment(QtCore.Qt.AlignCenter)
        self.label.setObjectName("label")
        self.START = QtWidgets.QPushButton(Form)
        self.START.setGeometry(QtCore.QRect(30, 230, 201, 51))
        self.START.setStyleSheet("background-color: rgb(0, 75, 0);color:rgb(255, 255, 255);\n"
"font: 75 20pt \"MS Shell Dlg 2\";")
        self.START.setCheckable(False)
        self.START.setObjectName("START")
        
        self.START.clicked.connect(self.startclicked)
        
        self.STOP = QtWidgets.QPushButton(Form)
        self.STOP.setGeometry(QtCore.QRect(30, 310, 201, 51))
        self.STOP.setStyleSheet("color:rgb(255, 255, 255);\n"
"background-color: rgb(189, 0, 0);\n"
"font: 75 20pt \"MS Shell Dlg 2\";")
        self.STOP.setCheckable(False)
        self.STOP.setObjectName("STOP")
        
        self.STOP.clicked.connect(self.stopclicked)
        
        self.imgLabel = QtWidgets.QLabel(Form)
        self.imgLabel.setGeometry(QtCore.QRect(300, 160, 741, 561))
        self.imgLabel.setStyleSheet("color:rgb(255, 255, 255);background-color: rgb(179, 179, 179);\n"
"font: 75 30pt \"MS Shell Dlg 2\";")
        self.imgLabel.setAlignment(QtCore.Qt.AlignCenter)
        self.imgLabel.setObjectName("imgLabel")
        self.label_2 = QtWidgets.QLabel(Form)
        self.label_2.setGeometry(QtCore.QRect(1100,260, 200,30))
        self.label_2.setStyleSheet("font: 75 22pt \"MS Shell Dlg 2\";color:rgb(255, 255, 255);\n"
"background-color:rgb(0, 65, 95);")
        self.label_2.setAlignment(QtCore.Qt.AlignCenter)
        self.label_2.setObjectName("label_2")
        self.label_3 = QtWidgets.QLabel(Form)
        self.label_3.setGeometry(QtCore.QRect(100, 760, 800, 41))
        self.label_3.setStyleSheet("font: 75 22pt \"MS Shell Dlg 2\";color:rgb(255, 255, 255);\n"
"background-color: rgb(0, 65, 95);")
        self.label_3.setAlignment(QtCore.Qt.AlignCenter)
        self.label_3.setObjectName("label_3")
        self.TEXT = QtWidgets.QLineEdit(Form)
        self.TEXT.setGeometry(QtCore.QRect(1120, 300, 201, 31))
        self.TEXT.setStyleSheet("color:rgb(0, 170, 0);\n"
"background-color: rgb(226, 226, 226);\n"
"font: 75 20pt \"MS Shell Dlg 2\";")
        self.TEXT.setText("")
        self.TEXT.setObjectName("TEXT")
        
        self.TEXT.setText("press start")
        
        self.PERCENT = QtWidgets.QLineEdit(Form)
        self.PERCENT.setGeometry(QtCore.QRect(900, 760, 140, 41))
        self.PERCENT.setStyleSheet("font: 75 22pt \"MS Shell Dlg 2\";color:rgb(255, 255, 255);\n"
"background-color:rgb(0, 65, 95);")
        self.PERCENT.setText("")
        self.PERCENT.setAlignment(QtCore.Qt.AlignCenter)
        self.PERCENT.setObjectName("PERCENT")
        self.SO = QtWidgets.QPushButton(Form)
        self.SO.setGeometry(QtCore.QRect(60, 540, 180, 41))
        self.SO.setStyleSheet("color:rgb(255, 255, 255);\n"
"font: 75 16pt \"MS Shell Dlg 2\";\n"
"background-color:rgb(0, 65, 95);")
        self.SO.setObjectName("SO")
        
        self.SO.clicked.connect(self.saved)
        
        self.label_5 = QtWidgets.QLabel(Form)
        self.label_5.setGeometry(QtCore.QRect(1100, 450, 300,51))
        self.label_5.setStyleSheet("font: 75 22pt \"MS Shell Dlg 2\";color:rgb(255, 255, 255);\n"
"background-color: rgb(0, 65, 95);")
        self.label_5.setAlignment(QtCore.Qt.AlignCenter)
        self.label_5.setObjectName("label_5")
        self.ACK = QtWidgets.QLineEdit(Form)
        self.ACK.setGeometry(QtCore.QRect(1120,510, 201, 31))
        self.ACK.setStyleSheet("color:rgb(0, 170, 0);\n"
"background-color: rgb(226, 226, 226);\n"
"font: 75 20pt \"MS Shell Dlg 2\";")
        self.ACK.setText("")
        self.ACK.setObjectName("ACK")
        self.label_7 = QtWidgets.QLabel(Form)
        self.label_7.setGeometry(QtCore.QRect(0, 0, 261, 121))
        self.label_7.setText("")
        self.label_7.setPixmap(QtGui.QPixmap("img.jpg"))
        self.label_7.setScaledContents(True)
        self.label_7.setObjectName("label_7")

        self.retranslateUi(Form)
        QtCore.QMetaObject.connectSlotsByName(Form)

    def retranslateUi(self, Form):
        _translate = QtCore.QCoreApplication.translate
        Form.setWindowTitle(_translate("Form", "Form"))
        self.label.setText(_translate("Form", "SOCIAL DISTANCE MONITORING"))
        self.START.setText(_translate("Form", "START"))
        self.STOP.setText(_translate("Form", "STOP"))
        self.imgLabel.setText(_translate("Form", "Live Monitor"))
        self.label_2.setText(_translate("Form", "STATUS"))
        self.label_3.setText(_translate("Form", "Percentage(%) of Social Distance Maintained"))
        self.SO.setText(_translate("Form", "Saved Output"))
        self.label_5.setText(_translate("Form", "Acknowledgement"))
        
    def saved(self):
        import webbrowser

        path = "output_videos"
        webbrowser.open(path)
        
    def startclicked(self):
        self.logic=1
        import numpy as np
        import cv2
        import imutils
        import os
        import time

        def CalculateDist(a,  b):
            dist = ((a[0] - b[0]) ** 2 + 550 / ((a[1] + b[1]) / 2) * (a[1] - b[1]) ** 2) ** 0.5
            calibration = (a[1] + b[1]) / 2       
            if 0 < dist < 1 * calibration:
                return True
            else:
                return False

        def Settingup(yolo):
            global net, ln, LABELS
            weights = os.path.sep.join([yolo, "yolov3.weights"])
            config = os.path.sep.join([yolo, "yolov3.cfg"])
            labelsPath = os.path.sep.join([yolo, "coco.names"])
            LABELS = open(labelsPath).read().strip().split("\n")  
            net = cv2.dnn.readNetFromDarknet(config, weights)
            ln = net.getLayerNames()
            ln = [ln[i[0] - 1] for i in net.getUnconnectedOutLayers()]

        def ProcessGivenFrame(image):
            global processedImg
            (H, W) = (None, None)
            frame = image.copy()
            if W is None or H is None:
                (H, W) = frame.shape[:2]
            blob = cv2.dnn.blobFromImage(frame, 1 / 255.0, (416, 416), swapRB=True, crop=False)
            net.setInput(blob)
            starttime = time.time()
            layerOutputs = net.forward(ln)
            stoptime = time.time()
            print("Video is Getting Processed at {:.4f} seconds per frame".format((stoptime-starttime))) 
            confidences = []
            outline = []

            for output in layerOutputs:
                for detection in output:
                    scores = detection[5:]
                    maxi_class = np.argmax(scores)
                    confidence = scores[maxi_class]
                    if LABELS[maxi_class] == "person":
                        if confidence > 0.5:
                            box = detection[0:4] * np.array([W, H, W, H])
                            (centerX, centerY, width, height) = box.astype("int")
                            x = int(centerX - (width / 2))
                            y = int(centerY - (height / 2))
                            outline.append([x, y, int(width), int(height)])
                            confidences.append(float(confidence))

            box_line = cv2.dnn.NMSBoxes(outline, confidences, 0.5, 0.3)
            one=0
            two=0
            if len(box_line) > 0:
                flat_box = box_line.flatten()
                pairs = []
                center = []
                status = [] 
                for i in flat_box:
                    (x, y) = (outline[i][0], outline[i][1])
                    (w, h) = (outline[i][2], outline[i][3])
                    center.append([int(x + w / 2), int(y + h / 2)])
                    status.append(False)

                for i in range(len(center)):
                    for j in range(len(center)):
                        close = CalculateDist(center[i], center[j])

                        if close:
                            pairs.append([center[i], center[j]])
                            status[i] = True
                            status[j] = True
                index = 0


                for i in flat_box:
                    (x, y) = (outline[i][0], outline[i][1])
                    (w, h) = (outline[i][2], outline[i][3])

                    if status[index] == True:
                        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 150), 2)
                        one=one+1

                    elif status[index] == False:
                        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
                        two=two+1
                    index += 1
                for h in pairs:
                    cv2.line(frame, tuple(h[0]), tuple(h[1]), (0, 0, 255), 2)
            processedImg = frame.copy()
            
            #check the percentage of social distence 
            sum1=one+two
            if sum1>0:
                if(two>0 or one>0):

                    total=(100*two)/sum1

                    tot = round(total, 2)
                    hey=str(tot)
                    
                    #resetting ack
                    self.ACK.setText("")
                    #for sending msg to email
                    global together 
                    together= (together+tot)/2
                    together2 = round(together, 2)

                    global start2 
                    if together2<=40 and start2.minute==datetime.now().minute :
                        start_time = datetime.now() 
                        together=0
                        from email.message import EmailMessage
                        import smtplib
                        def email_alert(subject,body,to):
                            msg=EmailMessage()
                            msg.set_content(body)
                            msg['subject']=subject
                            msg['to']=to

                            user="uthpalajshetty@gmail.com"
                            msg['from']=user
                            password="fatyycvozmdutehg"

                            server=smtplib.SMTP("smtp.gmail.com",587)
                            server.starttls()
                            server.login(user,password)
                            server.send_message(msg)

                            server.quit()
                            self.ACK.setText("mail sent")
                            print("msg sent")

                        if __name__=='__main__':
                            email_alert("social distance monitor","social distance is not maintained in this location","nayaksuraj78@gmail.com")
                            
                        start2=start_time + timedelta(minutes=1)
                            
                        start2=start_time + timedelta(minutes=1) 
                        
                        
                            
                    self.PERCENT.setText(hey+"%")
                else:
                    self.PERCENT.setText("N")
            else:
                self.PERCENT.setText("N")
            
        create = None
        frameno = 0
        yolo = "yolo-coco/"
        global n
        name="output"
        opname="output_videos/output1.mp4"
        
        if os.path.isfile(opname):
            n=n+1
            m=str(n)
            fname=name+m            
            opname="output_videos/"+fname+".mp4"
            
            
            
        cap = cv2.VideoCapture(0)
        
        
        
        time1 = time.time()
        while(True):

            ret, frame = cap.read()
            if not ret:
                break
            current_img = frame.copy()
            current_img = imutils.resize(current_img, width=480)
            video = current_img.shape
            frameno += 1

            if(frameno%2 == 0 or frameno == 1):
                Settingup(yolo)
                ProcessGivenFrame(current_img)
                Frame = processedImg
                if create is None:
                    fourcc = cv2.VideoWriter_fourcc(*'XVID')
                    create = cv2.VideoWriter(opname, fourcc, 10, (Frame.shape[1], Frame.shape[0]), True)
                    
                self.displayImage(Frame,1)
                cv2.waitKey()
                
                if(self.logic==1):
                    create.write(Frame)
                    self.TEXT.setText("started")
                
                if(self.logic==0):
                    self.TEXT.setText("stopped")
                    break
                    
                    
        cap.release()
        cv2.destroyAllWindows()  
        
    def stopclicked(self):
        self.logic=0
        
    def displayImage(self,img,window=1):
        qformat=QImage.Format_Indexed8
        
        if len(img.shape)==3:
            if (img.shape[2])==4:
                qformat= QImage.Format_RGBA888
            else:
                qformat= QImage.Format_RGB888
                
        img=QImage(img,img.shape[1],img.shape[0],qformat)
        img=img.rgbSwapped()
        img2=img.scaled(540, 540)
        self.imgLabel.setPixmap(QPixmap.fromImage(img2))
        
    
   



if __name__ == "__main__":
    import sys
    app = QtWidgets.QApplication(sys.argv)
    Form = QtWidgets.QWidget()
    ui = Ui_Form()
    ui.setupUi(Form)
    Form.show()
    sys.exit(app.exec_())
```
